```python
# importaremos a biblioteca pandas para carregar os dados

import pandas as pd
```

```python
# usaremos a funcao pd.read_csv para importar os dados externos para nosso notebook e 
# atribuiremos a uma variável chamada dados

# Parâmetros:
    # encoding => troca o codificador do arquivo csv
    # skipwors => ignora as 3 primeiras linhas ao carregar o arquivo
    # skipfooter => ignora as 12 últimas linhas ao carregar o arquivo
    # thousands => seta o separador de milhar para ponto (.)
    # decimal => seta o separador decimal para virgula (,)

dados = pd.read_csv('dados/A152918189_28_143_208.csv', 
                    encoding='ISO-8859-1', skiprows=3, skipfooter=12, 
                   sep=';', thousands='.', decimal=',')
```

# Importando outra versão dos dados

```python

dados = pd.read_csv('dados/A150850189_28_143_208.csv', 
                    encoding='ISO-8859-1', skiprows=3, skipfooter=12, 
                   sep=';', thousands='.', decimal=',')
```

```python
dados.head()
```

```python
# Nos dados acima podemos perceber que alguns meses estao sem valores informados. Esses dados não nos interessam.
# portanto, temos que removê-los. Mas como?

# Podemos usar o metodo de tirar a media de todas as colunas. As médias só serão apresentadas das colunas que
# tiverem todas as linhas preenchidas.

# O camando abaixo nos trará todas as médias das colunas preenchidas em todas as linhas. Dessa forma, podemos 
# capturar o nome das colunas, e criar uma lista com o nome dessas colunas.
dados.mean()

colunas_usaveis = dados.mean().index.to_list()


# precisamos incluir a coluna com os nomes da unidade da federacao
colunas_usaveis.insert(0, 'Unidade da Federação')

colunas_usaveis[:5]
```

```python
# Com a variavel colunas_usaveis, podemos criar um novo dataframe somente com essas colunas

dados_usaveis = dados[colunas_usaveis]
dados_usaveis.head()

```

```python
# Podemos agora, transformar a coluna Unidade da Federação em índice das linhas

dados_usaveis = dados_usaveis.set_index('Unidade da Federação')
dados_usaveis.head()
```

```python
# Agora vamos plotar as informações de todas as colunas
dados_usaveis.plot(figsize=(12,6))
```

```python
# No grafico acima, podemos perceber que não ficaram muito claras as informações.
# Queremos na verdade que os meses fiquem no eixo x, e as unidades no eixo y
# Podemos então transpor o dataframe, transformando as colunas em linhas e as linhas em colunas

dados_usaveis.T.head()
```

```python
# vamos agora criar novamente o gráfico com os dados transpostos

dados_usaveis.T.plot(figsize=(12,5))
```

```python
# No grafico acima, percebemos que no final tem uma variação muito grande da linha, o que indica que há algo errado
# se mostrarmos as ultimas colunas do grafico para averiguar, veremos que tem a coluna Total, que não pode fazer
# parte do gráfico

dados_usaveis.iloc[:,-3:].head()
```

```python
# Precisamos agora remover essa coluna.
# Parametros:
    # Nome da coluna
    # Informa que a coluna está no eixo x
    # Inplace = True já modifica o dataframe sem precisar redefini-lo
dados_usaveis.drop(['Total'], axis=1, inplace=True)
```

```python
# Após as alterações vamos plotar novamente o gráfico
axis = dados_usaveis.T.plot(figsize=(12,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
axis.legend(bbox_to_anchor=(1.3, 1.1))
plt.xlabel('período')
plt.ylabel('valor')
plt.title('Valor por mês das Unidades da Federação')
plt.show()
```

```python
dados.head()
```

```python
# Para melhorar a aparência do gráfico, podemos plotar somente 5 unidades da federação de forma aleatória
# a funcao sample() faz exatamente isso.

dados_usaveis.sample(n=5)

# Na analise de dados temos que reproduzir a aleatoriedade, pois dessa forma podemos reproduzir experimentos
# Podemos usar a biblioteca numpy para esse trabalho para criar uma semente (seed) que será o valor inicial
# para criar um numero aleatorio

import numpy as np

np.random.seed(524387)
dados_usaveis.sample(n=7)
```

```python
np.random.seed(524387)
axis = dados_usaveis.sample(n=7).T.plot(figsize=(12,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
axis.legend(bbox_to_anchor=(1.3, 1.1))
plt.ylabel('Gasto')
plt.xlabel('Período')
plt.title('Gastos mensais por Unidade da Federação')
plt.show()
```

```python
# vamos agora incluir novamente a coluna total, somando todos os valores por unidade de federacao

dados_usaveis['Total'] = dados_usaveis.sum(axis=1)
```

```python
dados_usaveis.head()
```

```python
# Podemos ordenar nosso conjunto de dados por total de gastos, do maior para o menor

dados_usaveis.sort_values('Total', ascending=False)
```

```python
# Vamos separar os 7 estados criados anteriormente em um novo dataframe
np.random.seed(524387)
dados_dos_7_estados = dados_usaveis.sample(7)
```

```python
# Agora vou incluir meu estado (Distrito Federal) nesse conjunto de dados 
dados_dos_7_estados

dados_dos_7_estados = dados_dos_7_estados.append(dados_usaveis.loc['53 Distrito Federal'])
```

```python
dados_dos_7_estados
```

```python
# Para remover dados duplicados
dados_dos_7_estados = dados_dos_7_estados.drop_duplicates()
```

```python
dados_dos_7_estados
```

```python
# Precisamos incluir uma coluna que informa a região de cada unidade federativa
dic_regiao = {'1': 'Norte', '2':'Nordeste', '3':'Sudeste', '4':'Sul', '5':'Centro-Oeste'}

def incluirRegiao(uf):
    codRegiao = uf.split(' ')[0][0][:1]
    return dic_regiao[codRegiao]
    

```

```python
#Testando a função

incluirRegiao('53 Distrito Federal')
```

```python
# Criando a coluna região no dataframe

dados_usaveis['Regiao'] = dados_usaveis.index.map(incluirRegiao)
```

```python
dados_usaveis
```

```python
dados_usaveis
```

```python
# vamos agora ordenar nosso dataframe pelo total, como fizemos anteriormente, só que agora atribuindo a um novo
# dataframe

ordenados_por_total = dados_usaveis.sort_values(by='Total', ascending=False)
```

```python
# o que pretendíamos era somente deixar nossos dados ordenados por total, portanto, agora podemos novamene
# excluir a coluna total

ordenados_por_total = ordenados_por_total.drop('Total', axis=1)
ordenados_por_total.head()
```

```python
ordenados_por_total.iloc[:,:-1].head().T.plot(figsize=(10,6))
```

```python
# Podemos observar que temos um pico de dados. Provavelmente são dados represados que se somaram nesse período
# Esses dados não nos interessam nesse momento, portanto iremos retirá-lo

# para remover essas colunas, podemos fazer um slice nas colunas que serão retiradas
colunas_interessadas = ordenados_por_total.columns[6:]
```

```python
ordenados_por_total = ordenados_por_total[colunas_interessadas]

# vamos plotar novamente o grafico
ordenados_por_total.iloc[:,:-1].head().T.plot(figsize=(10,6))
```

```python
# Podemos plotar os dados de uma determinada região do Brasil. Como criamos a coluna região, podemos filtrar por ela

ordenados_por_total.query("Regiao == 'Centro-Oeste'").iloc[:,:-1].T.plot(figsize=(12,6))
```
