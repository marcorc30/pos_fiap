## Análise exploratória de dados

```python
# importaremos a biblioteca pandas para carregar os dados

import pandas as pd
```

```python
# usaremos a funcao pd.read_csv para importar os dados externos para nosso notebook e 
# atribuiremos a uma variável chamada dados

# Parâmetros:
    # encoding => troca o codificador do arquivo csv
    # skipwors => ignora as 3 primeiras linhas ao carregar o arquivo
    # skipfooter => ignora as 12 últimas linhas ao carregar o arquivo
    # thousands => seta o separador de milhar para ponto (.)
    # decimal => seta o separador decimal para virgula (,)

dados = pd.read_csv('pos-datascience-analise-e-exploracao-de-dados-Dados/A152918189_28_143_208.csv', 
                    encoding='ISO-8859-1', skiprows=3, skipfooter=12, 
                   sep=';', thousands='.', decimal=',')
```

```python
# ler as 5 primeiras linhas
dados.head()
```

```python
# lear as 5 ultimas linhas
dados.tail()
```

```python
# informacoes sobre o dataframe
dados.info()
```

```python
# verificar a media de cada coluna do nosso dataframe 
dados.mean()
```

```python
# observe que no resultado acima, o resultado está sendo apresentado em notação científica, o que não serve
# no nosso caso. No entanto, podemos alterar a configuração global do pandas.

pd.options.display.float_format = '{:.2f}'.format
```
